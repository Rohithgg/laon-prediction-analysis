some more data cleaning and setting so that it can be easy to predict as we converted all string as int or boolean or float

```python
# data modeling 
train_df_labels = train["LoanAmount"].copy()
train_df = train.drop("LoanAmount",axis = 1)
```

```python
# Joining the Gender dummies to the DataFrame
gender_dummies = pd.get_dummies(train_df['Gender'], prefix='Gender')
train_df = train_df.join(gender_dummies).drop("Gender", axis=1)

# Renaming columns
train_df.rename(columns={'Female': 'Female_df1', 'Male': 'Male_df1'}, inplace=True)
train_df.rename(columns={'Female': 'Female_df2', 'Male': 'Male_df2'}, inplace=True)
```

```python
print(type(train_df))
```

```python
train_df = train_df.join(pd.get_dummies(train.Education)).drop("Education",axis=1)
```

```python
train_df = train_df.join(pd.get_dummies(train.Self_Employed)).drop("Self_Employed",axis=1)
```

```python
train_df = train_df.drop("Property_Area",axis = 1)
```

```python
train_df = train_df.join(pd.get_dummies(train.Loan_Status)).drop("Loan_Status",axis=1)
```

```python
train_df = train_df.join(pd.get_dummies(train_df['Married'],prefix='married')).drop("Married",axis=1)
```

```python
train_df = train_df.drop("N",axis=1)
```

```python
train_df = train_df.rename(columns={"Y":"Inloan"})
```

now we made a easy data to understand and loan_id was not requried,hence removed and as dependents has few srtings in it lets make it 3.5 meaning 3+ 

```python
# Replace '3+' with 3.5 in the 'Dependents' column
train_df['Dependents'] = train_df['Dependents'].replace('3+', 3.5)
```

```python
train_df = train_df.drop("Loan_ID",axis=1)
```

```python
train_df = train_df.rename(columns={'Yes':'self_employed'})
train_df = train_df.rename(columns={'No':'un_employed'})
```

```python

from sklearn.model_selection import cross_validate
from sklearn.metrics import confusion_matrix
# A sample toy binary classification dataset
X, y = datasets.make_classification(n_classes=2, random_state=0)
svm = LinearSVC(dual="auto", random_state=0)
def confusion_matrix_scorer(clf, X, y):
    y_pred = clf.predict(X)
    cm = confusion_matrix(y, y_pred)
    return {'tn': cm[0, 0], 'fp': cm[0, 1],
            'fn': cm[1, 0], 'tp': cm[1, 1]}
cv_results = cross_validate(svm, X, y, cv=5,
                            scoring=confusion_matrix_scorer)
# Getting the test set true positive scores
print(cv_results['test_tp'])
# Getting the test set false negative scores
print(cv_results['test_fn'])
```
